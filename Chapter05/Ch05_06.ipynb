```python
import datetime
import os
import pandas as pd

def get_df(subdir, file):
    
    year_season = subdir[-6:]
    year = int(year_season.split('S')[0])
    season = int(year_season.split('S')[1])
    
    stock_id = file.replace('.html', '')

    df = pd.read_html(os.path.join(subdir, file))[-1]
    
    df = pd.DataFrame({
        '會計項目': df[df.columns[0]],
        '金額': df[df.columns[1]]
    })

    df.drop_duplicates(subset='會計項目', keep='last', inplace=True)
    df.dropna(inplace=True)
    
    df = df.set_index('會計項目').T

    df['證券代號'] = stock_id
    df['日期'] = pd.to_datetime(season_to_date(year, season))
    
    df = df.set_index(['證券代號', '日期'])
        
    return df
```

```python
def get_df(subdir, file, cumulative_income_sheet=False):
    
    year_season = subdir[-6:]
    year = int(year_season.split('S')[0])
    season = int(year_season.split('S')[1])
    
    stock_id = file.replace('.html', '')

    df = pd.read_html(os.path.join(subdir, file))[-1]
    
    if cumulative_income_sheet and season in [2, 3]:
        df = pd.DataFrame({
            '會計項目': df[df.columns[0]],
            '金額': df[df.columns[5]]
        })
    else:
        df = pd.DataFrame({
            '會計項目': df[df.columns[0]],
            '金額': df[df.columns[1]]
        })
    
    df.drop_duplicates(subset='會計項目', keep='last', inplace=True)
    df.dropna(inplace=True)
    
    df = df.set_index('會計項目').T 

    df['證券代號'] = stock_id
    df['日期'] = pd.to_datetime(season_to_date(year, season))
    
    df = df.set_index(['證券代號', '日期'])
    
    if 'income_statement' in subdir and 'S4' in subdir:
        s3_df = get_df(subdir.replace('S4', 'S3'), file, cumulative_income_sheet=True)
        df.iloc[0] = df.iloc[0].sub(s3_df.iloc[0], fill_value=0)    
        
    return df
```

```python
import sqlite3

def save_to_db(table_name, new_df):
    
    connection = sqlite3.connect('data.db')
    
    try:
        df = pd.read_sql('select * from {}'.format(table_name), connection, parse_dates=['日期'], index_col=['證券代號', '日期'])
    except:
        df = pd.DataFrame()
    
    combined_df = df.append(new_df, sort=False) 
    
    final_df = combined_df.reset_index().drop_duplicates(subset=['證券代號', '日期'], keep='last').set_index(['證券代號', '日期']).sort_index()
    
    final_df.to_sql(table_name, connection, if_exists='replace')
    
    connection.close()
    
    excel_file = os.path.join('data',
                              'excel_files',
                              '{}.xlsx'.format(table_name))
    
    os.makedirs(os.path.dirname(excel_file), exist_ok=True)
    
    final_df.to_excel(excel_file)   
    
    return final_df
```

```python
folders = {
    'balance_sheet': os.path.join('data', 'financial_statement', 'balance_sheet'),
    'income_statement': os.path.join('data', 'financial_statement', 'income_statement'),
    'cash_flow': os.path.join('data', 'financial_statement', 'cash_flow')
}

for folder_name, path in folders.items():

    main_df = pd.DataFrame()
    
    for subdir, dirs, files in os.walk(folders[folder_name]):
        
        print('正在處理 {}.'.format(subdir))
        
        for file in files:      
            
            if file.endswith('.html'):
                try:
                    df = get_df(subdir, file)  
                except Exception as err:
                    print(subdir, file, err)
                else:
                    main_df = main_df.append(df, sort=False)
                    
    save_to_db(folder_name, main_df)
```
