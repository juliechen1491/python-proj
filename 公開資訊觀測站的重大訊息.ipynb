```
from bs4 import BeautifulSoup
import requests

headers = {
    'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'
}

# 公開資訊觀測站的重大訊息
main_url = 'https://mops.twse.com.tw/mops/web/t05sr01_1'
main_response = requests.get(main_url, headers=headers)
main_soup = BeautifulSoup(main_response.text, 'lxml')

# 取得表格
table = main_soup.select_one('table.hasBorder')

# 取得第一筆資料
tr = table.select('tr')[1]

# 取得詳情資料中的參數
document_values = tr.select('td')[-1].input['onclick'].split(';')

SEQ_NO = document_values[0].split('=')[-1].replace("'", '')
SPOKE_TIME = document_values[1].split('=')[-1].replace("'", '')
SPOKE_DATE = document_values[2].split('=')[-1].replace("'", '')
COMPANY_ID = document_values[3].split('=')[-1].replace("'", '')
skey = document_values[4].split('=')[-1].replace("'", '')

print('SEQ_NO', SEQ_NO,
      '\nSPOKE_TIME', SPOKE_TIME,
      '\nSPOKE_DATE', SPOKE_DATE, 
      '\nCOMPANY_ID', COMPANY_ID,
      '\nskey', skey)
```

```
url = 'https://mops.twse.com.tw/mops/web/ajax_t05sr01_1'

payload = {
    'encodeURIComponent': '1',
    'TYPEK': 'all',
    'step': 1,
    'firstin': 'true',
    'skey': skey,
    'COMPANY_ID': COMPANY_ID,
    'SPOKE_DATE': SPOKE_DATE,
    'SPOKE_TIME': SPOKE_TIME,
    'SEQ_NO': SEQ_NO
}

response = requests.get(url, params=payload, headers=headers)
soup = BeautifulSoup(response.text, 'lxml')
response.text
```

```
from requests.models import PreparedRequest

req = PreparedRequest()
req.prepare_url(url, payload)
req.url
```
