# Yahoo 財金新聞
Yahoo財金新聞-房地產：https://tw.news.yahoo.com/real-estate

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

site_url = 'https://tw.news.yahoo.com'

url = site_url + '/real-estate'

headers = {
    'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36'
}

response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text, 'lxml')
```

```python
li_tags = soup.findAll("li", {"class": "js-stream-content"})

title_list = []
paragraph_list = []
link_list = []

for news in li_tags:

    if 'http' in news.a.get('href'):
        continue
        
    title = news.h3.text
    paragraph = news.p.text
    link = site_url + news.a.get('href')
    
    print(title)
    print(paragraph)
    print(link)
    print('-'*20)
    
    title_list.append(title)
    paragraph_list.append(paragraph)
    link_list.append(link)
    
```

```python
df = pd.DataFrame({
    '標題': title_list,
    '內容': paragraph_list,
    '連結': link_list
})

df
```

```python
df.to_csv('最新房地產新聞.csv', encoding='utf-8')
```
